```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
ls1 = [12345,'Anwar','Developer',3]
ls2 = [12346,'Sara','Manager',6]
ls3 = [12347,'Ahmed','Developer',5]
ls = [ls1,ls2,ls3]
df = pd.DataFrame(ls,columns=['ID', 'Name', 'Role','Experience'])
plt.hist(df['Role'])
plt.xlabel('Role')
plt.ylabel('Number of empolyee')
plt.title('Number of employees in each Role')
plt.show()

plt.plot(df['Name'],df['Experience'])
plt.ylim(0,10)
plt.yticks(range(0,10,1))
plt.xlabel('Name')
plt.ylabel('Years')
plt.title("Employee's experience")
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
df.groupby('Role').count()
```

```python
Role = ['Developer','Manager']
co = [2,1]

plt.pie(co, labels= Role, autopct='%1.1f%%', startangle=90, shadow=True, radius=1, explode = [0,0]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df['Role'], df['Experience'], marker = 'o', s = 50, alpha =0.4)
plt.title("The experience for employee's in each role")
plt.xlabel('Role')
plt.ylabel('Years')
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(df['Role'], df['Experience'], marker = 'o', s = 50, alpha =0.4)
plt.title("The experience for employee's in each role",loc = 'left', c = 'r')
plt.xlabel('Role', fontsize = 16)
plt.ylabel('Years', fontsize = 16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot([0,0.5,2,0.7,3], 'r')
plt.plot([3,3,4,6,5], 'b')
plt.plot([0.2,2.3,4,4.5,6], 'c')
plt.legend(['Red','Blue','Cyan'],loc = 'upper left');
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots')

#first
plt.subplot(1,3,1) #(number of rows, number of columns, number of plot)
plt.pie(co, labels= Role, autopct='%1.1f%%')
plt.title('Role');

#second
plt.subplot(1,3,2)
plt.scatter(df['Role'], df['Experience'])
plt.title("The experience for employee's in each role")
plt.xlabel('Role')
plt.ylabel('Years');

#third 
plt.subplot(1,3,3)
plt.hist(df['Role'])
plt.xlabel('Role')
plt.ylabel('Number of empolyee')
plt.title('Number of employees in each Role');
```
